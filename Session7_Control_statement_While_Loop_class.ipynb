<font color = "#0073eb" size="5px"><b><u> Continue statement :</font></b><br></u>
 The `continue statement` we can stop the current iteration of the loop, and continue with the next.


```python
n=1
while n<=10:
  if n==5:
    n+=1
    continue
  print(n)
  n+=1
```

<font color = "#0073eb" size="5px"><b><u> While Loop with else</font></b><br></u>

* `else` block runs when **while condition becomes False** (not when `break` is used).

ðŸ“Œ Example:





```python
n=1
while n<=5:
  print(n)
  n+=1
else:
  print("done with the loop")
```

* If `break` is used, `else` will **not execute**.

ðŸ“Œ Example:





```python
num=1
while num<=10:
  if num==5:
    break
  print(num)
  num+=1
else:
  print("done with the loop")
print("I'm out")
```

```python
n=1
while n<=10:
  if n%2==0:
    n+=1
    continue
  print(n)
  n+=1
else:
  print("done with the loop")
print("I'm out")
```

## **6. Putting It All Together**

* Combine `while`, `else`, `break`, and `continue`.

ðŸ“Œ Example: Guess the number game.





* create 2 variable secret variable hold the secret number and "Guess" which intialized with 0
* run the while loop until guess!=secret
* get the guess number from the user
* check the guess nimber `<0` if yes print -ve number try again and continue the loop
* check whether guess==secret print you are correct break the loop
* else try again
* while loop -else "Game over"


```python
secret=56
guess=0
while guess!=secret:
  guess=int(input("Guess the number:"))
  if guess<0:
    print("It is a -ve number try again")
    continue
  elif secret==guess:
    print("You are correct")
    break
  else:
    print("you are wrong try again!")
else:
  print("Game over")
```

## **8. Practice Exercises**

1. Print numbers from 1 to 20, but stop when number is 13.
2. Print only odd numbers from 1 to 15 using `continue`.
3. Write a program that keeps asking the user to enter a password until correct. Use `break`.
4. Create a countdown timer from 10 to 1, and when finished print `"Happy New Year!"` using `else`.


```python

#Print numbers from 1 to 20, but stop when number is 13.

num = 1

while num <= 20:
    if num == 13:
        break
    print(num)
    num += 1
```

```python
# Print only odd numbers from 1 to 15 using continue.

num = 1

while num <= 15:
    if num % 2 == 0:   # even number
        num += 1
        continue       # skip even numbers
    print(num)
    num += 1
```

```python

# Write a program that keeps asking the user to enter a password until correct. Use break.

correct_password = "python123"

while True:
    pwd = input("Enter password: ")

    if pwd == correct_password:
        print("Access Granted!")
        break
    else:
        print("Wrong password. Try again.")
```

```python
# Create a countdown timer from 10 to 1, and when finished print "Happy New Year!" using else.

num = 11

while num >= 1:
    print(num)
    num -= 1
else:
    print("Happy New Year!")
```
