```python
#Strings (Concatenation, Slicing and Other methods) :

#1. String Concatenation:

#Write a Python program that takes two strings i.e string 1 “Hello ”, string 2 get name as input from the user and concatenates them together.

#Answer
string1 = "Hello"
string2 = input("Enter Your Name:")
result = string1 +" "+ string2
print(result)

#Display the concatenated string as the output.

#Sample Output: Enter your Name: Zara Hello Zara

#Now concatenate string 3 “, welcome to Python programming” to the existing string and display the output string.

string3 = ", welcome to Python programming"

final_string = result + " "+ string3

print(final_string)

#Sample Output: Hello Zara, welcome to Python programming
```

```python
#2. String Slicing and Indexing:
#Write a Python program using the above concatenated string as input and performs the following tasks:
string1 = "Hello"
name = input("Enter Your Name: ")
string2 = ", welcome to Python programming"
full_string = string1 + " "+ name + string2
print(full_string)

print("\nFull String:", full_string)

#a. Print the first character of the string.

print("a. First character of the string :", full_string[0])

#b. Print the last character of the string.

print("b. Last character of the string :", full_string[-1])

#c. Print the first 5 characters of the string.

print("c. First 5 characters of the string:", full_string[:5])

#d. Print the last 11 characters of the string.

print("d. Last 11 characters of the string:", full_string[-11:])

#e. Print the string in reverse.

print("e. String in reverse:", full_string[::-1])

#f. Use slicing and print the word “Python” from the existing string.

start_index = full_string.find("Python")

end_index = start_index + len("Python")

print("f. Word 'Python':", full_string[start_index:end_index])


```

```python
#3. String Methods: Write a Python program that takes a string, strM = “Python beginner tutorial” and perform the following tasks:

strM = "Python beginner tutorial"

print("\n Original String:", strM)

#a. Convert the sentence to uppercase.

print("\n a.Uppercase:",strM.upper())

#b. Convert the sentence to lowercase.

print("\n b.Lowercase:",strM.lower())

#c. Use Capitalize and return the sentence to the original input form.

print("\n c.Capitalize:",strM.capitalize())

#d. Count the total number of occurrences of character ‘t’ in the string.

print("\n d.Occurrences of character ‘t’ in the string:",strM.count('t'))

#e. Replace all occurrences of “Python” with “Machine Learning” in the input string strM = “Python beginner tutorial”

new_string = strM.replace ("Python","Machine Learning")

print("\n e.After Replacement: ", new_string)
```

```python
#Tuples (Creation, Modification and Access) :

#Create 1st tuple with values -> (10, 20, 30), 2nd tuple with values -> (40, 50, 60):

t1 = (10, 20, 30)
t2 = (40, 50, 60)

print("\n Tuple1: ", t1)
print("\n Tuple2: ", t2)

#a. Concatenate the two tuples and store it in “t_combine”

t_combine = t1 + t2

print("\n a. Concatenated of the two tuples:", t_combine)


#b. Repeat the elements of “t_combine” 3 times

t_repeat = t_combine * 3

print("\n b. Repeat the elements of “t_combine” 3 times:", t_repeat)

#c. Access the 3rd element from “t_combine”

print("\n 3rd Element: ", t_combine[2])

#d. Access the first three elements from “t_combine”

print("\n First three elements: ", t_combine[:3])

#e. Access the last three elements from “t_combine”

print("\n Last three elements: ", t_combine[-3:])
```
